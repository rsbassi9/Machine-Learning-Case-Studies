```python
#Drop the 4 rows with null values in the 'age' category
dataset = dataset[pd.notnull(dataset['age'])]

#Because the nulls are so many in credit_score and rewards_earned,
#we can remove these from our analysis
dataset = dataset.drop(columns = ['credit_score', 'rewards_earned'])
```

```python
## Histograms

#creat new dataset that only contains numeric data columns. Those that are
#non-numeric have no use in histograms. churn is our response variable.
dataset2 = dataset.drop(columns = ['user', 'churn'])

#Now, we creat a for loop to plot a histogram for every column (i)
fig = plt.figure(figsize=(15, 12))
plt.suptitle('Histograms of Numerical Columns', fontsize=20)
for i in range(1, dataset2.shape[1] + 1):
#plot the histograms in a 6x5 matrix of plots
    plt.subplot(6, 5, i)
    f = plt.gca()
    
#we dont want to overwhelm the plots, so get rid of y-axis values
    f.axes.get_yaxis().set_visible(False)
    
#give each a heading based on the column it references (i-1 in python indexing)
    f.set_title(dataset2.columns.values[i - 1])

#the vals variable allows us to create as many bins as we want in each subplot
    vals = np.size(dataset2.iloc[:, i - 1].unique())
    
    plt.hist(dataset2.iloc[:, i - 1], bins=vals, color='#3F5D7D')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
## Pie Plots for binary data
#we first set the new dataset to be representative only of the binary data
dataset2 = dataset[['housing', 'is_referred', 'app_downloaded',
                    'web_user', 'app_web_user', 'ios_user',
                    'android_user', 'registered_phones', 'payment_type',
                    'waiting_4_loan', 'cancelled_loan',
                    'received_loan', 'rejected_loan', 'zodiac_sign',
                    'left_for_two_month_plus', 'left_for_one_month', 'is_referred']]

#Same code as that for histograms, but slight changes 
fig = plt.figure(figsize=(15, 12))
plt.suptitle('Pie Chart Distributions', fontsize=20)
for i in range(1, dataset2.shape[1] + 1):
    plt.subplot(6, 3, i)
    f = plt.gca()
    f.axes.get_yaxis().set_visible(False)
    f.set_title(dataset2.columns.values[i - 1])
   
    #values are the ditributions of the columns in question(and all their rows)
    #but we want the distributions to be percentages, so we normalize the data
    values = dataset2.iloc[:, i - 1].value_counts(normalize = True).values
    index = dataset2.iloc[:, i - 1].value_counts(normalize = True).index
    plt.pie(values, labels = index, autopct='%1.1f%%')
    
    #remove the x and y labels as they dont exist for pie chart
    plt.axis('equal')
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
## Correlation of our individual variables with our Response Variable, churn

#this gives us a feel of which variables will be stronger predictors when building our model

#the correlation plot requires our fields to be numerical, so remove 
#categorical fields. Keep binary ones.

dataset.drop(columns = ['churn', 'user','housing', 'payment_type',
                          'zodiac_sign']
    ).corrwith(dataset.churn).plot.bar(figsize=(20,10),
              title = 'Correlation with Response variable',
              fontsize = 15, rot = 45,
              grid = True)

#Small bar = small correlation and vice versa
#downwards = -ve (more likely to stay)
#upwards = = +ve (more likely to churn)

##as we see below, age is -vely correlated. The smaller the age, the more 
#likely the response variable will be 1 (churn) and so on.

##credit card taken shows +ve correlation. This is weird since it means that people
#that received credit cards are more likely to churn. Maybe theyre not happy
#with the credit card features?

##High recommendations means less churn

##The more applications you begin, the more likely you stay with the firm
# aka, churn = 0

##Android users more likely to stay than iOS users
```

```python
## Correlation Matrix 

#allows us to visualize the above bar plot between the individual and response variable
sn.set(style="white")

# Compute the correlation matrix. Remove fields that are not useful for matrix
corr = dataset.drop(columns = ['user', 'churn']).corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(18, 15))

# Generate a custom diverging colormap.
#highest red = strongest relationship in matrix
#negative correlation = blue
cmap = sn.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sn.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

#As below, we see that there is only a positive correlation HIGH of 0.25, so we can
#keep any strongly correlated values(such as cc_application_begin)

#also,even though app_web_user doesnt show high correlation, we know it's a dependent, 
#correlated variable since app_web_user is only 1 if app_downloaded and 
#web_user are 1. Since it is a function of other fields(dependednt variable)
#we can get rid of it.
```

```python
# Removing Correlated Fields
dataset = dataset.drop(columns = ['app_web_user'])
```
